# Bonus Task

### Sentiment Analysis Expectations:
Given that the text is about key events of the twentieth century, I expect the sentiment analysis to show:
- **Low polarity** scores, as historical events often involve neutral or factual descriptions. However, events like wars or economic crises might result in negative scores.
- **Low to moderate subjectivity**, since the text might contain both factual statements (objective) and subjective interpretations (depending on the source).

Overall, I expect the text to have a slightly negative polarity and moderately subjective language.

```python
from textblob import TextBlob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('key_events_20th_century.txt', 'r', encoding='utf-8') as file: 
    text = file.read()
```

```python
# Break the text into sentences or paragrahs for more granular sentiment analysis
blob = TextBlob(text)
sentences = blob.sentences  # Tokenizing the text into sentences
```

```python
# Perform sentiment analysis (polarity and subjectivity) for each sentence/paragraph
sentiment_data = {
    "Sentence": [str(sentence) for sentence in sentences],
    "Polarity": [sentence.sentiment.polarity for sentence in sentences],
    "Subjectivity": [sentence.sentiment.subjectivity for sentence in sentences]
}
```

```python
# Convert the sentiment data into a DataFrame
df_sentiment = pd.DataFrame(sentiment_data)
```

```python
df_sentiment.head()
```

## Creating Plots to Visualize Sentiment Scores:

### Plotting Polarity and Subjectivity Scores

```python
# Plot the distribution of Polarity scores
plt.figure(figsize=(10, 5))
plt.hist(df_sentiment['Polarity'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Polarity Scores')
plt.xlabel('Polarity')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the distribution of Subjectivity scores
plt.figure(figsize=(10, 5))
plt.hist(df_sentiment['Subjectivity'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Subjectivity Scores')
plt.xlabel('Subjectivity')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of Polarity vs Subjectivity
plt.figure(figsize=(10, 5))
plt.scatter(df_sentiment['Polarity'], df_sentiment['Subjectivity'], alpha=0.5, color='purple')
plt.title('Polarity vs Subjectivity')
plt.xlabel('Polarity')
plt.ylabel('Subjectivity')
plt.grid(True)
plt.show()
```

### Sentiment Analysis Findings:
- **Polarity**: The polarity distribution shows a slight skew towards neutral or negative values, which matches my expectation, considering that many twentieth-century events were conflicts or crises.
- **Subjectivity**: The subjectivity distribution is moderate, suggesting a balance between factual reporting and subjective interpretation.
- **Polarity vs Subjectivity**: The scatter plot shows that sentences with high subjectivity tend to have more extreme polarity values, which is typical in historical narratives that blend fact and opinion.

Overall, my expectations about polarity being low and subjectivity being moderate were largely correct.
