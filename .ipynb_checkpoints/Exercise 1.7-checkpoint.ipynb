```python
import pandas as pd
import numpy as np
import networkx as nx
import os
import matplotlib.pyplot as plt
from pyvis.network import Network
import seaborn as sns
import pyvis 
```

```python
relationship_df = pd.read_csv('relationship_df.csv')
```

```python
# Create a graph from a pandas dataframe

G = nx.from_pandas_edgelist(relationship_df,
                            source="source",
                            target="target",
                            edge_attr="value",
                            create_using=nx.Graph())
```

```python
plt.figure(figsize = (10,10))
pos = nx.kamada_kawai_layout (G)
nx.draw(G, with_labels = True, node_color = 'red', edge_cmap = plt.cm.Blues, pos = pos)
```

```python
# Define net object

net = Network()
net = Network(notebook = True,width="900px", height="800px", bgcolor='#222222', font_color='white')
```

```python
# Define nodes from the G object
node_degree = dict(G.degree)
```

```python
# Setting up node size attributes
nx.set_node_attributes(G, node_degree, 'size')
net.from_nx(G)
net.show_buttons(filter_= True)
net.repulsion()
net.show("20thCentury.html")
```

```python
from cdlib import algorithms
coms = algorithms.leiden(G)
```

```python
new_com = coms.to_node_community_map()
```

```python
new_com
```

```python
# Put the output from new_com into a dictionary

dict_com = {k:v[0] for k, v in new_com.items()}
```

```python
# Rerun network graph with the communities as an argument

nx.set_node_attributes(G, dict_com, 'group')
com_net = Network(notebook=True, width="1300px", height="700px", bgcolor='#222222', font_color='white')
com_net.from_nx(G)
com_net.show_buttons(filter_=True)
com_net.repulsion()
com_net.show("20thCentury_communities_leiden.html")
```

### Partition Analysis and Historical Context:

1. **Red Community (Central Node: Russia)**:
   - Russia is the central node, representing its significant influence during **World War II** and the **Cold War**. Russia's connections with countries like **Germany**, **France**, and **Italy** reflect key historical events such as alliances and conflicts during the **World Wars** and later in the Cold War era.

2. **Blue Community**:
   - The blue cluster includes **Western European countries** like **France**, **UK**, and **Belgium**, symbolizing the alliances formed through **NATO** and the **European Union** post-World War II. 

3. **Yellow and Orange Communities**:
   - These communities reflect **Eastern European** and **Central European** countries, highlighting their roles during the Cold War.

4. **Green Community**:
   - The green cluster represents the **Balkan countries** and other regional players, likely reflecting post-Cold War conflicts such as the breakup of Yugoslavia and related tensions during the late 20th century.

5. **Purple Community**:
   - This cluster may represent countries like **China** and others that were more distanced from direct European conflicts but still influential globally, particularly during the second half of the century.

### Conclusions:
- The partitioning aligns with major geopolitical groupings from the **World Wars** to the **Cold War**. Russia plays a central role, with its influence extending across various European nations.
- Overall, the network graph provides an insightful visualization of the geopolitical landscape of the 20th century, showing how countries were connected during key historical events.

```python
# Degree centrality

degree_dict = nx.degree_centrality(G)
degree_dict
```

```python
degree_df = pd.DataFrame(list(degree_dict.items()), columns=['Country', 'Degree Centrality'])
```

```python
degree_df = degree_df.sort_values(by='Degree Centrality', ascending=False)
```

```python
plt.figure(figsize=(10,6))
plt.barh(degree_df['Country'], degree_df['Degree Centrality'], color='skyblue')
plt.xlabel('Degree Centrality')
plt.ylabel('Country')
plt.title('Degree Centrality of Countries')
plt.gca().invert_yaxis()  
plt.show()
```

```python
closeness_dict = nx.closeness_centrality(G)
```

```python
closeness_df = pd.DataFrame(list(closeness_dict.items()), columns=['Country', 'Closeness Centrality'])
```

```python
closeness_df = closeness_df.sort_values(by='Closeness Centrality', ascending=False)
```

```python
plt.figure(figsize=(10,6))
plt.barh(closeness_df['Country'], closeness_df['Closeness Centrality'], color='lightcoral')
plt.xlabel('Closeness Centrality')
plt.ylabel('Country')
plt.title('Closeness Centrality of Countries')
plt.gca().invert_yaxis()  
plt.show()
```

```python
betweenness_dict = nx.betweenness_centrality(G)
```

```python
betweenness_df = pd.DataFrame(list(betweenness_dict.items()), columns=['Country', 'Betweenness Centrality'])
```

```python
betweenness_df = betweenness_df.sort_values(by='Betweenness Centrality', ascending=False)
```

```python
plt.figure(figsize=(10,6))
plt.barh(betweenness_df['Country'], betweenness_df['Betweenness Centrality'], color='lightblue')
plt.xlabel('Betweenness Centrality')
plt.ylabel('Country')
plt.title('Betweenness Centrality of Countries')
plt.gca().invert_yaxis()  
plt.show()
```

### Centrality Analysis Observations

#### Degree Centrality
- **Germany** has the highest degree centrality, meaning it is the most connected country in this network, having direct ties with many others.
- **Russia** and **France** follow Germany, showing significant connections, which aligns with their historical roles and influence.
- **Poland** and **Romania** also have noticeable degrees of centrality, indicating their regional importance.
- Countries like **Lebanon** and **Mozambique** show the lowest degree centrality, with minimal direct connections to other nations.

#### Closeness Centrality
- **Germany** again tops the chart for closeness centrality, signifying its central position in the network. It can spread information quickly through the network, being close to other countries.
- **Russia** and **France** also score high, reflecting their global influence and capacity to interact quickly with other countries.
- Interestingly, even countries like **Libya** and **Morocco** have relatively high closeness centrality, which may point to their strategic positioning within regional networks.
- **Mozambique** and **Lebanon** are again at the bottom, indicating that they are more peripheral in this network.

#### Betweenness Centrality
- **Germany** scores the highest in betweenness centrality by a large margin. This indicates that Germany acts as a bridge between many countries and is crucial for facilitating interactions between otherwise disconnected parts of the network.
- **Philippines** and **Cambodia** surprisingly have higher betweenness scores, which could point to their roles in specific geopolitical networks.
- **France** and **Russia** are still prominent but not as dominant in betweenness as in other measures.
- Peripheral countries like **Lebanon**, **Thailand**, and **Luxembourg** have low betweenness centrality, meaning they don't frequently serve as bridges in the network.

These centrality measures highlight **Germanys** dominant position in this network across all metrics, while other countries exhibit varied levels of influence and connectivity. Countries with high closeness and degree centrality can quickly disseminate information or maintain widespread connections, while those with high betweenness centrality, like the Philippines, play critical roles as intermediaries.
