```python
country_entities = []
```

```python
for sentence in doc.sents:  
    for ent in sentence.ents:  
        if ent.label_ == "GPE" and ent.text in countries:  
            country_entities.append({
                "Sentence": sentence.text.strip(),
                "Country": ent.text
            })
```

```python
for entity in country_entities:
    print(f"Sentence: {entity['Sentence']}\n - Country: {entity['Country']}\n")
```
